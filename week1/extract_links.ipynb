```python
import requests
from pyquery import PyQuery
import numpy as np
import re
```

```python
r= requests.get("http://www.sina.com")
```

```python
html = r.content.decode("utf-8")
```

```python
doc = PyQuery(html)
```

```python
a_tags = doc("a")
```

```python
links = [PyQuery(a).attr("href") for a in a_tags if PyQuery(a).attr("href") != None and re.match("https?://.*",PyQuery(a).attr("href"))]
```
